```python
# Import Libraries
# %matplotlib inline makes your outputs appear and be stored within the notebook

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set() # for plot styling
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(12,6)
import os
```

```python
# Load Data Set

pd.set_option('display.width',5000)
os.chdir("C:\\Users\\HP\\Downloads\\Prajwal_UK\\Prajwal_UK2\\")
df=pd.read_csv('driver-data.csv')
display(df)
```

```python
# Create Independent Variable (x)

x=df.iloc[:,[1,2]].values
print(x)
```

```python
# Elbow Method

from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++')
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    display(wcss)
```

```python
# Elbow Method Graph

plt.plot(range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
# Model with two clusters

kmeans=KMeans(n_clusters=2)
df_analyze=df.drop('id',axis=1)
kmeans.fit(df_analyze)
```

```python
# CLuster Central Points

display(kmeans.cluster_centers_)
```

```python
# Display cluster names

print(kmeans.labels_)
print(len(kmeans.labels_))
```

```python
# Record Count by cluster

print(type(kmeans.labels_))
unique,counts=np.unique(kmeans.labels_,return_counts=True)
print(dict(zip(unique,counts)))
```

```python
# Plot the graph

df_analyze['cluster'] = kmeans.labels_
sns.set_style('whitegrid')
#aspect (the ratio of the width to the height) 
# palette = colour 
sns.lmplot(data=df_analyze,x= 'mean_dist_day',y= 'mean_over_speed_perc', hue='cluster',
           palette=['blue','green'],aspect=2,fit_reg=False)
plt.show()
```

```python
# Save the result as .xls   

Final = pd.concat([df,pd.DataFrame(kmeans.labels_, columns = ['Cluster'])],axis =1)
Final.to_excel("C:\\Users\\HP\\Downloads\\Prajwal_UK\\Prajwal_UK2\\driver-data_2_Cluster.xlsx")
display (Final)
```

```python
# Create model with 4 clusters    

kmeans =  KMeans (n_clusters =  4 , init = 'k-means++')
kmeans = kmeans.fit(x)
y_kmeans = kmeans.predict(x)
print (y_kmeans)
```

```python
# Plot the graph  

unique, counts = np.unique(kmeans.labels_, return_counts=True)
print(dict(zip(unique, counts)))
plt.scatter (x[y_kmeans == 0,0],x[y_kmeans == 0,1] ,s = 100, c='r', label = 'Cluster 1')
plt.scatter (x[y_kmeans == 1,0],x[y_kmeans == 1,1] ,s = 100, c='b', label = 'Cluster 2')
plt.scatter (x[y_kmeans == 2,0],x[y_kmeans == 2,1] ,s = 100, c='g', label = 'Cluster 3')
plt.scatter (x[y_kmeans == 3,0],x[y_kmeans == 3,1] ,s = 100, c='c', label = 'Cluster 4')
plt.legend()
plt.show()
```

```python
# Save the result as .xls   

Final = pd.concat([df,pd.DataFrame(y_kmeans, columns = ['Cluster'])],axis =1)
Final.to_excel("C:\\Users\\HP\\Downloads\\Prajwal_UK\\Prajwal_UK2\\driver-data_4_Cluster.xlsx")
display(Final)
```
